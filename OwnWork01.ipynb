```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df,train_size = 0.7, test_size = 0.3, shuffle = False)
```

```python
from keras.models import Sequential
from keras.layers import Activation
from tensorflow.keras.layers import LSTM,Input, Dense, Dropout

length_of_sequences = df_train.shape[1]
hidden_neurons = 128
in_out_neurons = 1

model = Sequential()
model.add(LSTM(hidden_neurons, batch_input_shape=(None, length_of_sequences, in_out_neurons), return_sequences=False))
model.add(Dropout(0.25))#念のためDropout
model.add(Dense(1))  
model.add(Activation("linear")) 
```

```python
model.compile(loss="mean_squared_error", optimizer="adam")
history = model.fit(df_train, batch_size=128, epochs=60, validation_split=0.2)
```
